Aquí se utiliza la IA para crear la lista de compras en base a la comida que el usuario desea a preparase. Tambien se debe añadir la cantidad de porciones que se necesitaria.

```python
context = "Ahora eres un cocinero profesional y ayudas a las personas a usar las mejores recetas para sus comidas"

# Bucle para hacer preguntas múltiples
while True:
    # Pregunta sobre la comida a preparar
    comida = input("Ingresa la comida que quieres preparar:")
    porciones = input("Ingresa la cantidad de porciones (1 por persona) que quieres preparar:")
    
    # Combinación del contexto y la pregunta
    prompt = f'Realiza una lista de ingredientes en formato JSON para preparar la siguiente comida: {comida}. Cantidad de porciones: {porciones} '
        
    conversation = [
        {"role":"user","content":prompt}
        ]
    
    # Solicitud para obtener la respuesta del modelo
    response = openai.ChatCompletion.create(
        model='gpt-3.5-turbo',
        messages=conversation,
        max_tokens=500
    )

    # Obtiene y muestra la respuesta del modelo
    message = response.choices[0]['message']
    print("{}".format(message['content']))

    # Pregunta si se quiere hacer otra pregunta
    another_question = input("¿Quieres agregar otra comida? (Si/No): ").lower()

    # Condicional para decidir si continuar o salir del bucle
    if another_question != 'si':
        print("¡Gracias!")
        break
    break
```
